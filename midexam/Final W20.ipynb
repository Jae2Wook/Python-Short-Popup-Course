## Q1) Linear Algebra

Solve the system of equations belowusing NumPy's linear algebra function.

\begin{align*}
& x + 2y - 3z + 4w = 12 \\
& 2x + 2y - 2z + 3w = 10 \\
& x - y + z - 2w = -4 \\
& y + z = -1 \\
\end{align*}

 

```python
import numpy as np

A = np.array([[1,2,-3,4], [2,2,-2,3], [1,-1,1,-2], [0,1,1,0]])
C = np.array([12,10,-4,-1])
c = np.linalg.solve(A,C)
print("in the order of x, y, z, w values", c)
```

Suppose you wanted to find the average inflation rate over the five years in the data. You can't just take the average of the values in the data, inflation rates afffect the value of the price index as time goes by, compounding on each other. To find the average inflation rate over five years we need the level of the price index that generated the inflation data. To do this, you would need to convert each year to an index. 2012 will become 100 because that is easiest. Then you will multiply 100 by 1 + the inflation rate in 2013 to get 2013's index. Next, you will multiply 2013's index by 1 + 2014's inflation rate to get 2014's index, and so on. See the table below for a demonstration.

|Year            | Inflation Rate (Percent)       |   Equation |    Index |  
|-------------------------|-----------------------| -------------|----------|
| 2012     | 2.64 | N/A          | 100|

| 2015 |1.44 | $1.0144 * 104.68$ | 106.19 |
| 2016 | 2.0 | $1.02 * 106.19$ | 108.31|

$ \bigg(\dfrac{Index_{2016}}{Index_{2012}}\bigg) ^ {1/t} + 1$
